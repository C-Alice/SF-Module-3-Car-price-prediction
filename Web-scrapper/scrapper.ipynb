### Создание словаря с моделями автомобилей искомых марок
>### Creating a dictionary, which contains models of cars with required brand 

```python
import requests
import re
from bs4 import BeautifulSoup

catalogue = 'https://auto.ru/htmlsitemap/mark_model_catalog.html'
response = requests.get(catalogue)
page = BeautifulSoup(response.text, 'lxml')
links =[i.get('href') for i in page.find('div', class_='sitemap').find_all('a')]

cars_dict = {}
brands_list = ['MERCEDES', 'INFINITI', 'NISSAN', 'BMW', 'VOLKSWAGEN','MITSUBISHI', 
               'TOYOTA', 'VOLVO', 'SKODA', 'LEXUS', 'AUDI', 'HONDA','SUZUKI']
for link in links:
    brand = re.findall('cars/(.+?)/',i)[0].upper()
    if brand in brands_list:
        if brand not in cars_dict:
            cars_dict[brand] = []
        model = re.findall('cars/.+?/(.+?)/',i)[0]
        cars_dict[brand].append(model.upper())
```
